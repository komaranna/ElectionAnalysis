```python
def TablfyHtml(filename):
    data = []
    filehandle = open(filename,'r')
    line = filehandle.readline()
    while '</html>' not in line:
        line = filehandle.readline()
        if '<table' in line:
            tempTable = []
            while '</table>' not in line:
                line = filehandle.readline()
                if '<tr' in line:
                    tempRow = []
                    line = filehandle.readline()
                    while '</tr>' not in line:
                        tempRow.append(line)
                        line = filehandle.readline()
                    tempTable.append(tempRow)

            data.append(tempTable)
            
    return data


def GetVoterTurnout(data):
    KE = []
    LE = []
    ME = []
    NE = []
    AE = []
    FE = []
    BE = []
    GE = []
    C = []
    EE = []
    IE = []
    JE = []
    for table in data:
        for i in range(len(table)):
            row = table[i]
            for j in range(len(row)):
                if ('KE' in row[j]) and ('<th>' in row[j]):
                    KE.append(table[i+1][j])
                elif ('LE' in row[j]) and ('<th>' in row[j]):
                    LE.append(table[i+1][j])
                elif ('ME' in row[j]) and ('<th>' in row[j]):
                    ME.append(table[i+1][j])
                elif ('NE' in row[j]) and ('<th>' in row[j]):
                    NE.append(table[i+1][j])
                elif ('AE' in row[j]) and ('<th>' in row[j]):
                    AE.append(table[i+1][j])
                elif ('FE' in row[j]) and ('<th>' in row[j]):
                    FE.append(table[i+1][j])
                elif (('BOE' in row[j]) or ('BE' in row[j])) and ('<th>' in row[j]):
                    BE.append(table[i+1][j])
                elif ('GE' in row[j]) and ('<th>' in row[j]):
                    GE.append(table[i+1][j])
                elif ('C' in row[j]) and ('<th>' in row[j]):
                    C.append(table[i+1][j])
                elif ('EE' in row[j]) and ('<th>' in row[j]):
                    EE.append(table[i+1][j])
                elif ('IE' in row[j]) and ('<th>' in row[j]):
                    IE.append(table[i+1][j])
                elif ('JE' in row[j]) and ('<th>' in row[j]):
                    JE.append(table[i+1][j])
                    
    return KE,LE,ME,NE,AE,FE,BE,GE,C,EE,IE,JE


def GetPartyStatistics(data):
    jobbik = []
    fidesz = []
    lmp = []
    mszp = []
    other = []
    for table in data:
        for i in range(len(table)):
            row = table[i]
            for j in range(len(row)):
                if ('JOBBIK' in row[j]):
                    jobbik.append(table[i][j+1])
                elif ('FIDESZ' in row[j]):
                    fidesz.append(table[i][j+1])
                elif ('LMP' in row[j]):
                    lmp.append(table[i][j+1])
                elif ('MSZP' in row[j]):
                    mszp.append(table[i][j+1])
                    
    return jobbik,fidesz,lmp,mszp


def ExtractNumbers(strList):
    for i in range(len(strList)):
        strList[i] = strList[i].replace('<td style="text-align:right;">','')
        strList[i] = strList[i].replace(' ','')
        strList[i] = strList[i].replace('<td>','')
        strList[i] = strList[i].replace('</td>','')
        strList[i] = strList[i].replace('&nbsp;','')
        strList[i] = strList[i].replace('\n','')
        
        try:
            strList[i] = int(strList[i])
        except:
            j = 0
            splice = strList[i][0:4]
            while splice != '<br>':
                j += 1
                splice = strList[i][j:j+4]
            temp = strList[i][0:j]
            try:
                strList[i] = int(temp)
            except:
                strList[i] = temp
    if len(strList) > 1:
        return strList[0], strList[1]
    elif len(strList) == 0:
        return None
    else:
        return strList[0]
```

```python
import glob
import csv

def WriteToCSV(filename,AE,FE,KE,LE,ME,NE,BE,GE,C,EE,IE,JE,jobbik,fidesz,lmp,mszp):
    with open(filename, 'a') as csvfile:
        elwriter = csv.writer(csvfile, delimiter=',',lineterminator = '\n')
        elwriter.writerow([AE,FE,KE,LE,ME,NE,BE,GE,C,EE,IE,JE,jobbik[0],fidesz[0],lmp[0],mszp[0],jobbik[1],fidesz[1],lmp[1],mszp[1]])

        
def ListAllFiles(folder):
    allfiles = []
    filenames = glob.glob(folder + '/*.txt')
    for filename in filenames:
        temp = filename.split("\\")
        temp = temp[-2] + '/' + temp[-1]
        allfiles.append(temp)
    
    return allfiles


def ReadAndWriteData(iStr):
    if len(iStr) == 1:
        iStr = '0' + iStr
    allfiles = ListAllFiles(iStr)
    
    for filename in allfiles:
        data = TablfyHtml(filename)

        KE,LE,ME,NE,AE,FE,BE,GE,C,EE,IE,JE = GetVoterTurnout(data)
        KE = ExtractNumbers(KE)
        LE = ExtractNumbers(LE)
        ME = ExtractNumbers(ME)
        NE = ExtractNumbers(NE)
        AE = ExtractNumbers(AE)
        FE = ExtractNumbers(FE)
        BE = ExtractNumbers(BE)
        GE = ExtractNumbers(GE)
        C = ExtractNumbers(C)
        EE = ExtractNumbers(EE)
        IE = ExtractNumbers(IE)
        JE = ExtractNumbers(JE)
        
        jobbik,fidesz,lmp,mszp = GetPartyStatistics(data)
        jobbik = ExtractNumbers(jobbik)
        fidesz = ExtractNumbers(fidesz)
        lmp = ExtractNumbers(lmp)
        mszp = ExtractNumbers(mszp)
        
        WriteToCSV('electoral.csv',AE,FE,KE,LE,ME,NE,BE,GE,C,EE,IE,JE,jobbik,fidesz,lmp,mszp)
```
