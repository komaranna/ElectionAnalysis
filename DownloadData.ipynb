```python
import requests
```

```python
def ConvertParamsToString(MParam,TParam,SzKData):
    MString = str(MParam)
    if len(MString) == 1:
        MString = '0' + MString
    
    TString = str(TParam)
    if len(TString) == 1:
        TString = '00' + TString
    elif len(TString) == 2:
        TString = '0' + TString
        
    SzKString = str(SzKData)
    if len(SzKString) == 1:
        SzKString = '00' + SzKString
    elif len(SzKString) == 2:
        SzKString = '0' + SzKString
    
    return MString, TString, SzKString
```

```python
urlBase = 'http://www.valasztas.hu/dyn/pv14/szavossz/hu/'

r = requests.get(urlBase + 'M20/T001/szkjkv_001.html', allow_redirects=False)
MParam = 20
TParam = 1
SzKData = 1
MString,TString,SzKString = ConvertParamsToString(MParam,TParam,SzKData)

while r.status_code == 200:
    
    while r.status_code == 200:
        open(MString + '_' + TString + '_' + SzKString + '.txt', 'wb').write(r.content)
        SzKData += 1
        MString,TString,SzKString = ConvertParamsToString(MParam,TParam,SzKData)
        urlTemp = urlBase + 'M' +  MString + '/' + 'T' + TString + '/' + 'szkjkv_' + SzKString + '.html'
        r = requests.get(urlTemp, allow_redirects=False)
    
    print TString + ' finished.'
    SzKData = 1
    TParam += 1
    MString,TString,SzKString = ConvertParamsToString(MParam,TParam,SzKData)
    urlTemp = urlBase + 'M' +  MString + '/' + 'T' + TString + '/' + 'szkjkv_' + SzKString + '.html'
    r = requests.get(urlTemp, allow_redirects=False)
```
